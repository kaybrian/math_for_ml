```python
carbon = data['co2']

plt.style.use('seaborn-whitegrid')

plt.title("Distribution of carbon Dioxide")
plt.xlabel('CO2 Emissions (grams per mile)')
plt.ylabel('Frequency')


plt.tight_layout()
plt.hist(carbon,color = "lightsteelblue", bins=100)

plt.show()

plt.savefig('/content/drive/MyDrive/Colab Notebooks/Distribution_of_carbon_Dioxide.png')
```

```python
feScore = data['feScore']
plt.style.use('seaborn-whitegrid')


plt.title('Distribution of Fuel Efficiency Score')
plt.xlabel('Fuel Efficiency Score')
plt.ylabel('Frequency')


plt.tight_layout()
plt.hist(feScore, bins=10, facecolor = '#f3ce5a', edgecolor='#faecc1', linewidth=0.5)
plt.show()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Distribution_of_Fuel_Efficiency_Score.png')
```

```python
highway = data['highway']

plt.style.use('seaborn-whitegrid')

plt.xlabel('Highway Fuel Efficiency (MPG)')
plt.ylabel('Frequency')
plt.title('Distribution of Highway Fuel Efficiency')


plt.tight_layout()
plt.hist(highway, bins=30, facecolor = '#f3ce5a', edgecolor='#faecc1', linewidth=0.5)
plt.show()
plt.savefig('/content/drive/MyDrive/Colab Notebooks/Distribution_of_Highway_Fuel_Efficiency.png')
```
